```python
import requests
from pprint import pprint
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import time
import pandas as pd
import numpy as np
from textblob import TextBlob, Word
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords, wordnet
```

## 1) Write a function that will scrape book title and its personal page link from a given url.


```python
base_url = "http://books.toscrape.com/"
catalogue_url = base_url + "catalogue/"
pages_to_scrap = 5                            #I wrote 5 pages as was requered, 

def get_titles_and_urls(url):
    response = requests.get(url)
    page = response.content
    page = BeautifulSoup(page, "html.parser")
    title_html = page.find_all("img", class_="thumbnail") 
    titles_arr = [i.get("alt") for i in title_html]                 #to find titles
    url_arr = page.find_all("div", class_="image_container")
    url_arr = [catalogue_url + i.find("a").get("href") for i in url_arr]   #to find urls of the books
    

    return [url_arr, titles_arr]  #returns array

```

```python
urls = [catalogue_url + "page-" + str(i) + ".html" for i in range(1, pages_to_scrap + 1)]     #5 pages plus 1 for range
book_urls = []
book_titles = []

for i in urls:
    page_data = get_titles_and_urls(i)  #calling the function
    book_urls.extend(page_data[0])   #index 0 for url_arr...to extend all 5 pages' urls
    book_titles.extend(page_data[1]) #index 1 for titles_arr ...5 pages' titles
    time.sleep(3)


pprint(book_titles)
pprint(book_urls)
```

### 2) Write a function that will scrape the book description from a given book page url.


```python

def scrape_descr(url):
    response = requests.get(url)
    page = response.content
    page = BeautifulSoup(page, "html.parser")
    article = page.find("article", class_="product_page")
    #pprint(article)
    return article.findChild("p", recursive = False).get_text()   # to find the p of "direct(first level)" Child that has inside a description
        
book_descriptions = [scrape_descr(i) for i in book_urls]  #implement function for all book_urls
pprint(book_descriptions)
```

```python
book_descriptions_blob = [TextBlob(i) for i in book_descriptions]  #to make TextBlob 

for i in book_descriptions_blob:
    print(i.sentiment)                 #to find polaryty and subjectivity of each book description
    print("\n") 
```

```python
all_book_descriptions_string = " ".join(book_descriptions)  #with join I made all book descriptions a string
pprint(all_book_descriptions_string)
```

```python
#making textBlob
all_books_desc = TextBlob(all_book_descriptions_string).lower()
#singularization of words
all_books_desc_words = [i.singularize() for i in all_books_desc.words]
#removing stopwords
sw = stopwords.words('english')
sw.append("'")  #as they also were counted as words, I eliminated them
sw.append("’")  #the same
sw.append('”')
all_books_desc_clean = [i for i in all_books_desc_words if i not in sw]   #if words are not in sw,so they are "clean" words, take them)
pprint(all_books_desc_clean)
```

```python
#getting the most popular word
nltk.FreqDist(all_books_desc_clean).most_common(1)
```
